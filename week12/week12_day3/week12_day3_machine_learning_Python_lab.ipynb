```python
songs = pd.get_dummies(songs, drop_first = True)

songs.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = songs["target"]
X = songs.drop(columns = "target")
```

```python
model = LinearRegression()

model.fit(X, y)
```

```python
model.score(X, y)
```

```python
model.intercept_
```
